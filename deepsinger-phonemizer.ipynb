```python
filepath = "/content/drive/MyDrive/Colab/Research/Lyrics/"
from phonemizer import phonemize
import os
import tqdm
for i in tqdm.tqdm(os.listdir(filepath)):
  if not os.path.isfile(f"/content/drive/MyDrive/Colab/Research/phonemized/{i}"):
    with open(f"/content/drive/MyDrive/Colab/Research/Lyrics/{i}") as f:
      line = f.read()
      phonemized = phonemize(line, language='en-us', backend='espeak')
      with open(f"/content/drive/MyDrive/Colab/Research/phonemized/{i}", 'w', encoding="utf-8") as f:
        f.write(phonemized)
      # for line in f.readlines():
      #   print(line)
```
